```python
import json

with open('/Users/imgkkk574/Documents/studying/Code/psy/psy-consult/src/students_config_0718.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

print(type(data))
print(len(data))
# 472条真实学生配置 消融实验部分 采样30条作为训练集 5条作为测试集
```

```python
import random
# --- 随机采样 ---
sample_train = random.sample(data, 30)
remaining = [item for item in data if item not in sample_train]
sample_test = random.sample(remaining, 5)

# 写入 train.json
with open('train.json', 'w', encoding='utf-8') as f:
    json.dump(sample_train, f, indent=2, ensure_ascii=False)

# 写入 test.json
with open('test.json', 'w', encoding='utf-8') as f:
    json.dump(sample_test, f, indent=2, ensure_ascii=False)
```
